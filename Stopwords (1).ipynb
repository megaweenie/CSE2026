```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
def remove_stop_words(string):
  stop_words=set(stopwords.words('english'))
  words=string.split()
  filtered_words=[word for word in words if word.lower() not in stop_words]
  new_string=' '.join(filtered_words)
  return new_string
input_string="This is an example sentence to remove stopwords from."
result=remove_stop_words(input_string)
print("Original String:",input_string)
print("Modified String:",result)
```

```python
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
nltk.download('stopwords')
def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string
input_string = "This is an example sentence to remove stopwords from."
result = remove_stop_words(input_string)
stopwords_count = len(input_string.split()) - len(result.split())
labels = ['Stopwords', 'Non-stopwords']
values = [stopwords_count, len(result.split())]
plt.bar(labels, values, color=['red', 'blue'])
plt.xlabel('Type of Words')
plt.ylabel('Count')
plt.title('Count of Stopwords vs Non-stopwords')
plt.show()
print("Original String:", input_string)
print("Modified String:", result)
print("Number of stopwords:", stopwords_count)
```
