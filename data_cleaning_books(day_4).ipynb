```python
import pandas as pd

import numpy as np

df=pd.read_csv("/BL-Flickr-Images-Book (1).csv")
print(df.head(10))
df.shape
```

```python
to_drop=["Edition Statement","Former owner","Corporate Author","Corporate Contributors","Contributors","Issuance type","Shelfmarks"]
df.drop(to_drop,inplace=True,axis=1)
df.head()
```

```python
for col in df.columns:
  print(col)
```

```python
df.set_index("Identifier",inplace=True)
df.head()
```

```python
df["Date of Publication"].head(25)
```

```python
print("Removing of unwanted characters")
unwanted_characters=['[',',','-']
def clean_dates(items):
  dop=str(items.loc["Date of Publication"])
  if dop=="non" or dop[0]=="[":
    return np.NaN
  for character in unwanted_characters:
    if character in dop:
      character_index=dop.find(character)
      dop=dop[:character_index]
  return dop
df["Date of Publication"]=df.apply(clean_dates,axis=1)
df["Date of Publication"].head(25)
```

```python
print("cleaning the title of the book")
def clean_title(Title):
  if Title=="NaN":
    return "NaN"
  if Title[0]=="[":
    Title=Title[1:Title.find("}")]
  if "by"in Title:
    Title=Title[:Title.find("by")]
  if "By" in Title:
    Title=Title[:Title.find("By")]
  if "[" in Title:
    Title=Title[:Title.find("[")]
  Title=Title[:-2]
  Title=list(map(str.capitalize,Title.split()))
  return ''.join(Title)
df["Title"]=df["Title"].apply(clean_title)
df.head()
```
