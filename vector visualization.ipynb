```python
fig,ax1=plt.subplots(figsize=(14,8))
x=np.arange(0,2.2,0.2)
y=np.arange(0,2.2,0.2)
X,Y=np.meshgrid(x,y)
x_direct=np.cos(X)*Y
y_direct=np.sin(Y)*Y
n=-2
color=np.sqrt(((y_direct-n)/2)*2+((x_direct-n)/2)*2)
ax1.quiver(X,Y,x_direct,y_direct,color,alpha=0.8)
ax1.set_aspect('equal')
ax1.set_title('Meshgrid Function')
```
